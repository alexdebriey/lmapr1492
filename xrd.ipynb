```python
from pymatgen.io.cif import CifParser
parser = CifParser("mp984718.cif")
structure = parser.get_structures()[0]

from pymatgen.analysis.diffraction.xrd import XRDCalculator
c = XRDCalculator(1.54060, 0, None).get_pattern(structure, scaled=True, two_theta_range=(0, 90))

a = c.x[0]
b = c.x[1]
d= c.x[2]
e = c.x[3]
a2= c.y[0]
b2 = c.y[1]
d2 = c.y[2]
e2 = c.y[3]
a3 = c.hkls[0]
b3 = c.hkls[1]
d3 = c.hkls[2]
e3 = c.hkls[3]
a4 = c.d_hkls[0]
b4 = c.d_hkls[1]
d4 = c.d_hkls[2]
e4 = c.d_hkls[3]

print('The first peak is at:')
print('2theta =', a, '°')
print('Intensity =', a2, '(CPS)')
print(a3)
print()
print('The second peak is at:')
print('2theta =', b, '°')
print('Intensity =', b2, '(CPS)')
print(b3)
print('this peak is just near the first peak as showed on the following figure:')
f= XRDCalculator()
f.show_plot(structure)
print()
print('The third peak is at:')
print('2theta =', d, '°')
print('Intensity =', d2, '(CPS)')
print(d3)
print()
print()
print('Extra: The fourth peak is at:')
print('2theta =', e, '°')
print('Intensity =', e2, '(CPS)')
print(e3)
print()
print()
print('references: Pymatgen.org, MaterialsProject.org, Jupyterhub')
print('Credits to: Gian-Marco Riganese')
```
