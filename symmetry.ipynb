```python
from pymatgen.io.cif import CifParser
parser = CifParser("mp984718.cif")
structure = parser.get_structures()[0]


#1
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
symmetry[0] =  SpacegroupAnalyzer(structure).get_symmetry_operations(cartesian=False)
premier = symmetry[7].as_xyz_string()
print("Symmetry 1: mirror plane")
print("Operation:")
print(premier)
print("Image for atom B (0.33,0.67,0.37):")

from IPython.display import Image
a = Image("mirrorplane.PNG")
display(a)

#2
deuxième = symmetry[2].as_xyz_string()
print("Symmetry 2: C3 axis")
print("Operation:")
print(deuxième)
print("Image for atom As (0.33,0.67,0):")

from IPython.display import Image
b=Image("c3axis.PNG")
display(b)

#3
troisième = symmetry[11].as_xyz_string()
print("Symmetry 3: Mirror plane")
print("Operation:")
print(troisième)
print("Image for atom As (0.67,0.33,0.5):")

from IPython.display import Image
c=Image("mirrorplane2.PNG")
display(c)
```
